```python
import pandas as pd
```

```python
df = pd.read_csv("alcohol_consumption.tsv", sep = "[,\t]", engine = "python")
df.head()
```

```python
# Comprobamos los nulos:
df.isnull().sum()
```

```python
# Revisamos los nombres de las columnas:
df.columns
```

```python
# Quitamos espacios finales a las columnas:
df.columns = df.columns.str.strip()
df.head()
```

```python
# Comprobamos las categorÃ­as de "frequenc":
df["frequenc"].unique()
```

```python
# Nos quedamos sÃ³lo con "DAY", "MTH" y "WEEK":
df = df.drop(df[df["frequenc"] == "LT1M"].index).drop(df[df["frequenc"] == "NM12"].index).drop(df[df["frequenc"] == "NVR"].index).drop(df[df["frequenc"] == "NVR_NM12"].index)
df["frequenc"].unique()
```

```python
# Comprobamos los intervalos de "age":
df["age"].unique()
```

```python
# Eliminamos todos los intervalos que no vamos a usar en el estudio:
df = df.drop(df[df["age"] == "TOTAL"].index).drop(df[df["age"] == "Y15-29"].index).drop(df[df["age"] == "Y15-64"].index).drop(df[df["age"] == "Y18-24"].index).drop(df[df["age"] == "Y18-44"].index).drop(df[df["age"] == "Y18-64"].index).drop(df[df["age"] == "Y25-64"].index).drop(df[df["age"] == "Y45-54"].index).drop(df[df["age"] == "Y65-74"].index).drop(df[df["age"] == "Y55-64"].index).drop(df[df["age"] == "Y_GE18"].index).drop(df[df["age"] == "Y_GE65"].index).drop(df[df["age"] == "Y_GE75"].index)
df["age"].unique()
```

```python
# Comprobamos las categorÃ­as de "sex":
df["sex"].unique()
```

```python
# Eliminamos "T":
df = df.drop((df[df["sex"] == "T"]).index)
df["sex"].unique()
```

```python
# Comprobamos las categorÃ­as de "citizen"...
df["citizen"].unique()
```

```python
#... y vemos que aquella con info es "NAT":
df.groupby(by = "citizen").agg(sum)
```

```python
# Nos quedamos con sÃ³lo NAT:
df = df.drop((df[df["citizen"] == "EU28_FOR"]).index).drop((df[df["citizen"] == "FOR"]).index).drop((df[df["citizen"] == "NEU28_FOR"]).index)
df.head()
```

```python
# Reseteamos index:
df = df.reset_index()
df = df.drop("index", axis = 1)
df.head()
```

```python
# Comprobamos que las columnas "unit", "time\geo", "EU28", "citizen" no proporcionan info, las borramos:
df = df.drop(["unit", "time\geo", "EU28", "citizen"], axis = 1)
df.head()
```
