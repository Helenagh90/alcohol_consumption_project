```python
import pandas as pd
```

```python
df = pd.read_csv("alcohol_consumption.tsv", sep = "[,\t]", engine = "python")
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
df.head()
```

```python
df.groupby(by = "frequenc").agg(sum)
```

```python
df = df.drop(df[df["frequenc"] == "LT1M"].index).drop(df[df["frequenc"] == "NM12"].index).drop(df[df["frequenc"] == "NVR"].index).drop(df[df["frequenc"] == "NVR_NM12"].index)
df.groupby(by = "frequenc").agg(sum)
```

```python
df.groupby(by = "age").agg(sum)
```

```python
df = df.drop(df[df["age"] == "TOTAL"].index).drop(df[df["age"] == "Y15-29"].index).drop(df[df["age"] == "Y15-64"].index).drop(df[df["age"] == "Y18-24"].index).drop(df[df["age"] == "Y18-44"].index).drop(df[df["age"] == "Y18-64"].index).drop(df[df["age"] == "Y25-64"].index).drop(df[df["age"] == "Y45-54"].index).drop(df[df["age"] == "Y65-74"].index).drop(df[df["age"] == "Y55-64"].index).drop(df[df["age"] == "Y_GE18"].index).drop(df[df["age"] == "Y_GE65"].index).drop(df[df["age"] == "Y_GE75"].index)
df.groupby(by = "age").agg(sum)
```

```python
df.groupby(by = "sex").agg(sum)
```

```python
df = df.drop((df[df["sex"] == "T"]).index)
df.groupby(by = "sex").agg(sum)
```

```python
df.groupby(by = "citizen").agg(sum)
```

```python
df = df.drop((df[df["citizen"] == "EU28_FOR"]).index).drop((df[df["citizen"] == "FOR"]).index).drop((df[df["citizen"] == "NEU28_FOR"]).index)
df.groupby(by = "citizen").agg(sum)
```
